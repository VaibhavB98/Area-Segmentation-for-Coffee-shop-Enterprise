<h1> Part-3 Getting Coordinates

```python
# define a function to get coordinates
def get_latlng(neighborhood):
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Pune , India'.format(neighborhood))
        lat_lng_coords = g.latlng
    return lat_lng_coords
```

```python
coords = [ get_latlng(neighborhood) for neighborhood in Pune_df["Neighbourhoods"].tolist() ]
```

```python
coords
```

```python
 #create temporary dataframe to populate the coordinates into Latitude and Longitude
df_coords = pd.DataFrame(coords, columns=['Latitude', 'Longitude'])
```

```python
# merge the coordinates into the original dataframe
Pune_df['Latitude'] = df_coords['Latitude']
Pune_df['Longitude'] = df_coords['Longitude']
```

<h3>Analyzing Each Neighbourhood

```python
# one hot encoding
pun_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
pun_onehot['Neighbourhoods'] = venues_df['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [pun_onehot.columns[-1]] + list(pun_onehot.columns[:-1])
pun_onehot = pun_onehot[fixed_columns]

print(pun_onehot.shape)
pun_onehot.head()
```

```python
#Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category
pun_grouped = pun_onehot.groupby(["Neighbourhoods"]).mean().reset_index()

print(pun_grouped.shape)
pun_grouped
```

```python
len(pun_grouped[pun_grouped["Coffee Shop"] > 0])
```

<h4>Create a new DataFrame for Coffee Shop data only

```python
Pun_Coff = pun_grouped[["Neighbourhoods","Coffee Shop"]]
Pun_Coff.head()
```

<h1>Part-6 Clustering Neighborhoods on Basis of Occurence of coffee shops 
    in an area. 

```python
# set number of clusters
kclusters = 3

pun_clustering = Pun_Coff.drop(["Neighbourhoods"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(pun_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# create a new dataframe that includes the cluster label as well as the mean of occurence.
pun_merged = Pun_Coff.copy()

# add clustering labels
pun_merged["Cluster Labels"] = kmeans.labels_
```

```python
pun_merged.rename(columns={"Neighbourhoods": "Neighborhood"}, inplace=True)
pun_merged.head()
```

```python
# merge  dataframes to add latitude/longitude for each neighborhood
pun_merged = pun_merged.join(Pune_df.set_index("Neighbourhoods"), on="Neighborhood")

print(pun_merged.shape)
pun_merged.head() # check the last columns!
```

```python
print(pun_merged.shape)
pun_merged.sort_values(["Cluster Labels"], inplace=True)
pun_merged
```

<h1> Part-7 Visualizing Clusters.

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(pun_merged['Latitude'], pun_merged['Longitude'], pun_merged['Neighborhood'], pun_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
